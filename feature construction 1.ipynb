```python
import pandas as pd
import numpy as np 
import seaborn as sns

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
#demonstate the features construction by combining and spliting to extract the information from the dataset and 
#write the conclusion  of survial status of different salutation 
```

```python
df = pd.read_csv("train - train.csv")[['Age', 'Pclass', 'SibSp', 'Parch', 'Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
x = df.iloc[:, 0:4]
y = df.iloc[:, -1]
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), x,y, scoring="accuracy", cv = 20))
```

```python
#Applying the Features Construction 
```

```python
x['Family Size'] = x['SibSp'] + x['Parch'] + 1
```

```python
def myfun(num):
    if num == 1:
        #alone
        return 0
    elif num > 1 and num <= 4:
        #small family 
        return 1
    else:
        #large family 
        return 2 
```

```python
myfun(4)
```

```python
x['Family_type'] = x['Family Size'].apply(myfun)
```

```python
x.head()
```

```python
x.drop(columns = ['SibSp','Parch','Family Size'], inplace=True)
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), x,y, scoring="accuracy", cv = 20))
```
